```python
def dfs(start, goal, graph, vis=None):
    if vis is None:
        vis = set()

    print(start)

    if start == goal:
        return True

    vis.add(start)

    for child in graph[start]:
        if child not in vis:
            if dfs(child, goal, graph, vis):
                return True
    return False
graph={}
n=int(input("enter no of nodes:"))
for i in range(n):
    li=[]
    no=input("enter node:")
    ni=int(input("enter number of neighbours:"))
    for j in range(ni):
        k=input("enter neighbour neighbour node:")
        li.append(k)
    graph[no]=li
print(graph)
```
