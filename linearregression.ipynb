```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import random
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
np.random.seed(42)
df=pd.DataFrame({
    "year":np.array([1,2,3,4,2,7,9,8,10,11]),
    "sal":np.array([20000,25000,32000,35000,27000,38000,40000,42000,48000,51000])
})
df
```

```python
df.to_csv("salary_pred.csv",index=False)
df.head()
```

```python
X=df[["year"]]
y=df["sal"]
```

```python
X_test,X_train,y_test,y_train=train_test_split(X,y,test_size=0.3)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X)
```

```python
print("slope:",model.coef_)
print("intercept:",model.intercept_)
```

```python
plt.scatter(X,y,label='Actual Data',color='blue')
plt.plot(X,y_pred,label='Predicted Data',color='red')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.show()
```

```python
new=np.array([[2],[10],[8]])
pred=model.predict(new)
for i in range(len(new)):
    print(new[i][0],"year:",round(pred[i],2))
```

```python
y_pred=model.predict(X_test)
print("mean absolute error:",mean_absolute_error(y_test,y_pred))
print("mean squared error:",mean_squared_error(y_test,y_pred))
print("r2 score:",r2_score(y_test,y_pred))
```
